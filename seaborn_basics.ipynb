```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pydataset import data
```

```python
data() # shows all the data available
passenger = data('AirPassengers')
#data('AirPassengers', show_doc = True)
passenger.describe()
sns.relplot(x = 'time', y= 'AirPassengers',palette = 'bright', data = passenger, kind = 'line')
```

```python
sns.jointplot(x = 'time', y= 'AirPassengers', data = passenger, kind = 'resid')
```

```python
sns.violinplot(x = 'time', y= 'AirPassengers',data = passenger )
```

```python
sns.lineplot(x = 'time', y= 'AirPassengers',data = passenger)
```
